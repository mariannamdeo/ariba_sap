```python
import os
import pandas as pd
import datetime
import lxml 

lista_de_pastas = ["Contrato Legal Compras","Projeto de Sourcing","Solicitação Sourcing"]

for pasta in lista_de_pastas:
    print('>>> INICIO: ', datetime.datetime.now().strftime('%d-%b-%Y %H:%M:%S'), ' | ', pasta)
    #os.getenv("USERNAME"), instalar a biblioteca OS
    # Para trocar o caminho "Documents" if os.getenv('LANG') == "en_US.UTF-8" else "Documentos"

    # Caminho da pasta onde estão os arquivos .xls
    minha_matricula = 'F8091818'
    caminho_da_pasta = 'C:\\Users\\' + minha_matricula + '\\TIM\\Consolidation, Reporting & KPIs - Documents\\Data Management\\Projetos\\Acomp. de Contratos\\ARIBA\\INPUTs\\' + pasta

    # Lista para armazenar os DataFrames lidos
    listas_de_dados = []

    # Loop para ler todos os arquivos .xls da pasta
    for nome_arquivo in os.listdir(caminho_da_pasta):
        if nome_arquivo.endswith('.xls'):
            # Caminho completo do arquivo
            caminho_arquivo = os.path.join(caminho_da_pasta, nome_arquivo)
            
            # Ler o arquivo .xls
            
            df = pd.read_html(caminho_arquivo)
            # Adicionar o DataFrame à lista
            df_tabela = df[1]
            new_df_tabela = df_tabela.assign(NomeArquivo=nome_arquivo)
            listas_de_dados.append(new_df_tabela)

    # Concatenar todos os DataFrames da lista em um único DataFrame
    df_final = pd.concat(listas_de_dados, ignore_index=True)

    # Salvar cada um em um df diferente, que será limpo pelas funções
    # Opcional: Salvar o DataFrame final em um novo arquivo .csv ou .xls
    df_final.to_csv('C:\\Users\\' + minha_matricula + '\\TIM\\Consolidation, Reporting & KPIs - Documents\\Data Management\\Projetos\\Acomp. de Contratos\\ARIBA\\CSVs_Unificados\\' + pasta.replace(" ","_") + '.csv', index=False)
    print('>>> FIM: ', datetime.datetime.now().strftime('%d-%b-%Y %H:%M:%S'), ' | ', pasta)
    # Ou para salvar como .xls
    # df_final.to_excel('base_unificada.xlsx', index=False)
```
