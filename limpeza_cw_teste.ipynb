```python
import pandas as pd
import numpy as np

#FEITO!!!
```

```python
df_ariba1 = pd.read_csv('Contrato_Legal_Compras.csv', sep=',', encoding='latin1')
df_ariba1.head()

#Código do projeto,CNPJ/CPF do Fornecedor,Fornecedor - Fornecedor global,Contrato - Data de efetivação,Contrato - Data de vencimento,Baseado no projeto... - Código do projeto,Contrato-raiz - Código do projeto,Status do contrato,count(),sum(Valor do contrato),NomeArquivo
```

```python
df_ariba1 = df_ariba1.drop(['Contrato - Data de efetivaÃ§Ã£o', 'Contrato - Data de vencimento', 'Status do contrato', 'count()'], axis=1)
df_ariba1.head()
```

```python
print (df_ariba1.dtypes)
```

```python


df_ariba1['sum(Valor do contrato)'] = (df_ariba1['sum(Valor do contrato)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float).round(2))

#df_ariba1.info()
df_ariba1.head()
```

```python
df_ariba1[['NomeArquivo1', 'NomeArquivo2', 'NomeArquivo3', 'NomeArquivo4', 'NomeArquivo5']] = df_ariba1['NomeArquivo'].str.split('_', expand=True)
df_ariba1.head()
```

```python
df_ariba1 = df_ariba1.drop(['NomeArquivo','NomeArquivo2','NomeArquivo3','NomeArquivo4','NomeArquivo5'], axis=1)
df_ariba1 = df_ariba1.rename(columns={'NomeArquivo1': 'DATA_REFERENCIA'})

df_ariba1.head()
```

```python
df_ariba1['CNPJ/CPF do Fornecedor'] = (df_ariba1['CNPJ/CPF do Fornecedor'].astype(str).str.replace('.', '', regex=False).str.replace('-', '.', regex=False).str.replace('/', '', regex=False).str.replace("'", "", regex=False))

#df_ariba1.info()
df_ariba1.head(10)
```

```python

df_ariba1 = df_ariba1.rename(columns={'Baseado no projeto... - CÃ³digo do projeto': 'BASEADO_NO_PROJ', 'CÃ³digo do projeto': 'CW', 'CNPJ/CPF do Fornecedor':'CNPJ_CPF_FORNEC', 'Fornecedor - Fornecedor global': 'FORNECEDOR', 'sum(Valor do contrato)': 'VALOR_CONTRATO', 'Contrato-raiz - CÃ³digo do projeto': 'CW_RAIZ'})
df_ariba1.head()
```

```python
df_ariba1['BASEADO_NO_PROJ'] = df_ariba1['BASEADO_NO_PROJ'].replace('Irrestrito', '', regex=False)
df_ariba1.head()
```

```python
df_ariba1['BASEADO_NO_PROJ'] = df_ariba1['BASEADO_NO_PROJ'].astype(str)
df_ariba1['SR'] = np.where(df_ariba1['BASEADO_NO_PROJ'].str.startswith('SR'), df_ariba1['BASEADO_NO_PROJ'],'' )
df_ariba1['WS'] = np.where(df_ariba1['BASEADO_NO_PROJ'].str.startswith('WS'), df_ariba1['BASEADO_NO_PROJ'],'' )

df_ariba1.head(10)
```

```python
df_ariba1.sort_values(by='CW')
df_ariba1.head(10)
```

```python
#df_ariba1.info()

df_ariba1 = df_ariba1[df_ariba1['VALOR_CONTRATO']>0]
df_ariba1.head(10)
```

```python
df_ariba1['DATA_REFERENCIA'] = pd.to_datetime(df_ariba1['DATA_REFERENCIA'])
df_ariba1.head()
```

```python
df_ariba1 = df_ariba1.drop(['BASEADO_NO_PROJ'], axis=1)
df_ariba1.head()
```

```python
data_mais_recente = df_ariba1['DATA_REFERENCIA'].max() #coluna NomeArquivo1	

df_ariba1 = df_ariba1[df_ariba1['DATA_REFERENCIA'] == data_mais_recente]
df_ariba1.head()
```

```python
df_ariba1['CW_RAIZ'] = df_ariba1['CW_RAIZ'].replace("Irrestrito", "", regex=False)
df_ariba1.head()
```

```python
ariba_cw_reordenado = ['CW', 'CW_RAIZ', 'CNPJ_CPF_FORNEC', 'FORNECEDOR', 'VALOR_CONTRATO', 'DATA_REFERENCIA', 'SR', 'WS']
df_ariba1 = df_ariba1.loc[:, ariba_cw_reordenado] #todas as linhas e todas as colunas
df_ariba1.head()
```
