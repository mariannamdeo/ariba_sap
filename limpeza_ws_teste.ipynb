```python
import pandas as pd
import numpy as np

# FEITO!!!!
```

```python
df_ariba_ws = pd.read_csv('Projeto_de_Sourcing.csv', sep=',', encoding='latin1')
df_ariba_ws.head()
```

```python
df_ariba_ws = df_ariba_ws.drop(['[PSO] Area do comprador', 'count(SourcingProject)'], axis=1)
df_ariba_ws.head()
```

```python
colunas = ['[PSO]Data inicial (Data)', '[PSO]Data final (Data)', '[PSO]Baseado no projeto... (CÃ³digo do projeto)', '[PSO]Projeto raiz (CÃ³digo do projeto)', '[New FormulÃ¡rio de AprovaÃ§Ã£o]Fornecedor Selecionado (Nome do fornecedor (L1))']

df_ariba_ws[colunas] = df_ariba_ws[colunas].replace('Irrestrito', '', regex=False)
```

```python
#sum(Valor Valor Negociado Ano 2)	sum(Valor Valor Negociado Demais Anos)	sum(Valor Valor Negociado Total)

df_ariba_ws = df_ariba_ws.rename(columns = {'sum(Valor Valor Negociado Ano 1)':'VALOR_ANO_1', 
                                 'sum(Valor Valor Negociado Ano 2)':'VALOR_ANO_2',
                                 'sum(Valor Valor Negociado Demais Anos)':'VALOR_DEMAIS_ANOS',
                                 'sum(Valor Valor Negociado Total)':'VALOR_TOTAL'})
df_ariba_ws.head()
```

```python


troca_tipo = ['VALOR_TOTAL', 'VALOR_DEMAIS_ANOS', 'VALOR_ANO_2', 'VALOR_ANO_1']

df_ariba_ws[troca_tipo] = (df_ariba_ws [troca_tipo].astype(str).apply(lambda col: col.str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float).round(2)))

#df_ariba_ws.info()
df_ariba_ws.head()
```

```python
df_ariba_ws[['NomeArquivo_1', 'NomeArquivo2', 'NomeArquivo3', 'NomeArquivo4', 'NomeArquivo5']] = df_ariba_ws['NomeArquivo'].str.split('_', expand=True)
df_ariba_ws.head()
```

```python
df_ariba_ws = df_ariba_ws.drop(['NomeArquivo','NomeArquivo2', 'NomeArquivo3', 'NomeArquivo4', 'NomeArquivo5'], axis = 1)
df_ariba_ws.head()
```

```python
df_ariba_ws = df_ariba_ws.rename(columns= {'NomeArquivo.1':'DATA_REFERENCIA', 
'[New FormulÃ¡rio de AprovaÃ§Ã£o]Fornecedor Selecionado (Nome do fornecedor (L1))' :'NOME_FORNECEDOR', 
'[PSO] Unidade organizacional do comprador' : "UO_COMPRADOR", 
'[PSO]Data inicial (Data)' : 'DATA_INI', 
'[PSO]Data final (Data)' : 'DATA_FIM', 
'[PSO]Projeto (CÃ³digo do projeto)' : 'WS', 
'[PSO]Baseado no projeto... (CÃ³digo do projeto)' : 'BASEADO_NO_PROJ', 
'[PSO]Projeto raiz (CÃ³digo do projeto)' : 'PROJ_RAIZ'})

df_ariba_ws.head()
```

```python
df_ariba_ws['SR'] = np.where(
    df_ariba_ws['BASEADO_NO_PROJ'] == df_ariba_ws['PROJ_RAIZ'], df_ariba_ws['PROJ_RAIZ'],
    np.where(
        (df_ariba_ws['BASEADO_NO_PROJ'] == '') & (df_ariba_ws['PROJ_RAIZ'] != ''), df_ariba_ws['PROJ_RAIZ'],
        np.where(
            (df_ariba_ws['PROJ_RAIZ'] == '') & (df_ariba_ws['BASEADO_NO_PROJ'] != ""), df_ariba_ws['BASEADO_NO_PROJ'],
            np.where(
                (df_ariba_ws['BASEADO_NO_PROJ'] != df_ariba_ws['PROJ_RAIZ']) & 
                (df_ariba_ws['PROJ_RAIZ'] != '') & 
                (df_ariba_ws['BASEADO_NO_PROJ'] != ''), df_ariba_ws['PROJ_RAIZ'],
                ''))))

df_ariba_ws.head()
```

```python
df_ariba_ws[['DATA_INI', 'DATA_FIM']] = df_ariba_ws[['DATA_INI', 'DATA_FIM']].apply(pd.to_datetime, errors = 'coerce', dayfirst=True)

#df_ariba1['DATA_REFERENCIA'] = pd.to_datetime(df_ariba1['DATA_REFERENCIA'])
```

```python
df_ariba_ws = df_ariba_ws.drop(['BASEADO_NO_PROJ', 'PROJ_RAIZ'], axis=1)
```

```python
df_ariba_ws['NomeArquivo_1'] = pd.to_datetime(df_ariba_ws['NomeArquivo_1'], format = '%YYYY/%mm/%dd', errors = 'coerce')
df_ariba_ws.head()
```

```python
ariba_ws_reordenado = ['WS', 'SR', 'DATA_INI', 'DATA_FIM', 'UO_COMPRADOR','NOME_FORNECEDOR', 'VALOR_ANO_1', 'VALOR_ANO_2', 'VALOR_DEMAIS_ANOS', 'VALOR_TOTAL', 'NomeArquivo_1']
df_ariba_ws = df_ariba_ws.loc[:, ariba_ws_reordenado]
df_ariba_ws = df_ariba_ws.drop(['DATA_INI','DATA_FIM','UO_COMPRADOR'], axis=1)
df_ariba_ws.head()
```
